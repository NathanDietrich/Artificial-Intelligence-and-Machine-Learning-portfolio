```python
import pandas as pd
import numpy as np
import yfinance as yf
import requests
import matplotlib.pyplot as plt
from datetime import datetime, timedelta


# 1. GET TOP 50 STOCKS BY MARKET CAP
top_50_stocks = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "TSLA", "BRK-B", "META", "V", "UNH",
    "JNJ", "XOM", "WMT", "JPM", "LLY", "PG", "MA", "HD", "MRK", "AVGO",
    "CVX", "PEP", "ABBV", "KO", "ORCL", "COST", "BAC", "MCD", "PFE", "ADBE",
    "CRM", "ACN", "DIS", "TXN", "ABT", "VZ", "NFLX", "TMO", "NKE", "DHR",
    "BMY", "WFC", "NEE", "LIN", "PM", "INTC", "HON", "UPS", "MS", "AMD"
]
# In practice, you would programmatically derive this list.


# ---------------------------
# 2. GET TOP 50 CRYPTOCURRENCIES BY MARKET CAP
# ---------------------------
top_50_cryptos = [
    "BTC-USD", "ETH-USD", "BNB-USD", "XRP-USD", "USDC-USD", "ADA-USD", "DOGE-USD", "SOL-USD", "MATIC-USD", "DOT-USD",
    "LTC-USD", "SHIB-USD", "TRX-USD", "AVAX-USD", "UNI-USD", "LINK-USD", "XLM-USD", "ATOM-USD", "ETC-USD", "XMR-USD",
    "BCH-USD", "APE-USD", "NEAR-USD", "ALGO-USD", "FIL-USD", "VET-USD", "FTM-USD", "MANA-USD", "SAND-USD", "EGLD-USD",
    "HBAR-USD", "ICP-USD", "FLOW-USD", "FRAX-USD", "QNT-USD", "AAVE-USD", "EOS-USD", "THETA-USD", "XTZ-USD", "CRO-USD",
    "RPL-USD", "AXS-USD", "GRT-USD", "NEO-USD", "CHZ-USD", "LDO-USD", "WAVE-USD", "CAKE-USD", "GALA-USD", "KLAY-USD"
]



# ---------------------------
# 3. FETCH HISTORICAL DATA
# ---------------------------

def fetch_stock_data(ticker, period="1y"):
    # Using yfinance to get 1 year of data by default
    return yf.download(ticker, period=period)

def fetch_crypto_data(crypto_id, days=365):
    # CoinGecko historical data: last 365 days daily data
    url = f"https://api.coingecko.com/api/v3/coins/{crypto_id}/market_chart"
    params = {
        "vs_currency": "usd",
        "days": days
    }
    response = requests.get(url, params=params).json()

    # response['prices'] is a list of [timestamp, price]
    df = pd.DataFrame(response["prices"], columns=["timestamp", "price"])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('timestamp', inplace=True)
    df.rename(columns={"price": "Close"}, inplace=True)
    return df
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Top 50 stocks
top_50_stocks = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "TSLA", "BRK-B", "META", "V", "UNH",
    "JNJ", "XOM", "WMT", "JPM", "LLY", "PG", "MA", "HD", "MRK", "AVGO",
    "CVX", "PEP", "ABBV", "KO", "ORCL", "COST", "BAC", "MCD", "PFE", "ADBE",
    "CRM", "ACN", "DIS", "TXN", "ABT", "VZ", "NFLX", "TMO", "NKE", "DHR",
    "BMY", "WFC", "NEE", "LIN", "PM", "INTC", "HON", "UPS", "MS", "AMD"
]

# Example top 50 cryptos available on Yahoo Finance
top_50_cryptos = [
    "BTC-USD", "ETH-USD", "BNB-USD", "XRP-USD", "USDC-USD", "ADA-USD", "DOGE-USD", "SOL-USD", "MATIC-USD", "DOT-USD",
    "LTC-USD", "SHIB-USD", "TRX-USD", "AVAX-USD", "UNI-USD", "LINK-USD", "XLM-USD", "ATOM-USD", "ETC-USD", "XMR-USD",
    "BCH-USD", "APE-USD", "NEAR-USD", "ALGO-USD", "FIL-USD", "VET-USD", "FTM-USD", "MANA-USD", "SAND-USD", "EGLD-USD",
    "HBAR-USD", "ICP-USD", "FLOW-USD", "FRAX-USD", "QNT-USD", "AAVE-USD", "EOS-USD", "THETA-USD", "XTZ-USD", "CRO-USD",
    "RPL-USD", "AXS-USD", "GRT-USD", "NEO-USD", "CHZ-USD", "LDO-USD", "WAVES-USD", "CAKE-USD", "GALA-USD", "KLAY-USD"
]

all_assets = top_50_stocks + top_50_cryptos

data_dict = {}
print("Fetching data for all 100 assets...")
for asset in all_assets:
    print(f"Fetching data for {asset}...")
    df = yf.download(asset, period="1y")
    df = df.dropna()  # Drop any rows with NaN values
    if not df.empty:
        data_dict[asset] = df
    else:
        print(f"No data found for {asset} after dropping NaNs.")

# Create a 10x10 grid of subplots
fig, axes = plt.subplots(nrows=10, ncols=10, figsize=(30,30))
fig.suptitle("Top 50 Stocks and Top 50 Cryptos (1-Year Close Prices)", fontsize=20)
axes = axes.flatten()

for i, (asset, df) in enumerate(data_dict.items()):
    ax = axes[i]
    ax.plot(df['Close'], label=asset, alpha=0.7)
    ax.set_title(asset, fontsize=8)
    ax.tick_params(axis='x', labelrotation=45, labelsize=6)
    ax.tick_params(axis='y', labelsize=6)

# If fewer than 100 assets are available, remove extra subplots
for j in range(len(data_dict), 100):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
